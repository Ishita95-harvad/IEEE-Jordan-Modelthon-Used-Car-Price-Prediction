```python
import pandas as pd

# Feature Engineering

# 1. Car Age
current_year = 2025
df['Car_Age'] = current_year - df['Year']
df['Car_Age'] = df['Car_Age'].clip(lower=0) # Handle potential negative ages

# 2. Average Annual Mileage
df['Avg_Annual_Mileage'] = df['Kilometers'] / df['Car_Age']
df['Avg_Annual_Mileage'] = df['Avg_Annual_Mileage'].fillna(0) # Handle division by zero errors

# 3. Interaction Features (Example: Car Make and Price)
# One-hot encode 'Car Make'
car_make_dummies = pd.get_dummies(df['Car Make'], prefix='Car_Make')
df = pd.concat([df, car_make_dummies], axis=1)

# Create interaction features
for col in car_make_dummies.columns:
    df[f'{col}_x_Price'] = df[col] * df['Price']

# 4. Features from Options Columns
# Helper function to process options
def process_options(options_str):
    if pd.isna(options_str):
        return []
    return [option.strip() for option in options_str.split(',')]

options_cols = ['Interior Options', 'Exterior Options', 'Technology Options']
for col in options_cols:
    df[col] = df[col].apply(process_options)

# Create binary features for each option
all_options = []
for col in options_cols:
    all_options.extend([option for sublist in df[col] for option in sublist])
unique_options = set(all_options)

for option in unique_options:
    df[option] = df.apply(lambda row: 1 if option in row['Interior Options'] or option in row['Exterior Options'] or option in row['Technology Options'] else 0, axis=1)

display(df.head())
```

## Data cleaning

### Subtask:
Clean the data by handling missing values and addressing inconsistencies.


**Reasoning**:
Start cleaning the data by handling missing values, converting data types, and addressing inconsistencies as per the instructions.  This will involve imputing missing values, converting relevant columns to numeric types, and handling outliers and inconsistencies in categorical columns.



```python
import numpy as np

# 1. Missing Value Imputation
for col in ['Year', 'Kilometers', 'Engine Size (cc)', 'Number of Seats', 'Description_Score', 'Battery Capacity', 'Battery Range']:
    if df[col].isnull().sum() / len(df) > 0.6:
        print(f"Dropping column '{col}' due to high percentage of missing values.")
        df = df.drop(col, axis=1)
    elif df[col].dtype in ['int64', 'float64']:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])

# 2. Data Type Conversion
for col in ['Year', 'Kilometers', 'Engine Size (cc)', 'Number of Seats']:
    try:
        df[col] = df[col].astype(str).str.replace(r'[^\d.]', '', regex=True)
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].fillna(df[col].median())  # Impute with the median after conversion
    except:
        print(f"Unable to convert '{col}' to numeric, consider removing the column.")

# 3. Outlier Handling
for col in ['Year', 'Kilometers', 'Engine Size (cc)', 'Number of Seats']:
    if col in df.columns: #Check if column still exists after potential drops
      q1 = df[col].quantile(0.25)
      q3 = df[col].quantile(0.75)
      iqr = q3 - q1
      lower_bound = q1 - 1.5 * iqr
      upper_bound = q3 + 1.5 * iqr
      df[col] = np.clip(df[col], lower_bound, upper_bound)


# 4. Inconsistent Data (example for 'Car Make', expand for others)
df['Car Make'] = df['Car Make'].str.lower()  # Standardize to lowercase
df['Car Make'] = df['Car Make'].str.strip()  # Remove leading/trailing whitespace
display(df.head())
```
